```python
import csv
f = open("c:\sample\seoul.csv")
data = csv.reader(f)
header = next(data)
print(header)
result = []
for row in data:
    if row[-1] != "":
        if row[0].split("-")[1] == "08" and row[0].split("-")[2] == "15": # 1908-08-02 ->["1908", "08", "02"]
            result.append(float(row[-1]))
```
