```python
import csv
f = open("c:\sample\seoul.csv", "r")
data = csv.reader(f, delimiter = ",")
print(data)
header = next(data) # 첫번째 행 사용
maxt = []
for row in data:
    if (row[-1] != ""): # 공백이 있어서 float으로 바뀌지 않음
        maxt.append(float(row[-1]))
max_num = maxt[0]
for num in maxt:
    if(num > max_num):
        max_num = num
print("최고온도 : %f"  % max_num)

import matplotlib.pyplot as plt
plt.plot(maxt)
```
