```python
from bs4 import BeautifulSoup as bts
from urllib.request import urlopen
```

```python
html = urlopen("https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC")
print(html)
```

```python
page = bts(html, "html.parser")
page
```

```python
toctitle = page.find("div", class_ = "toctitle")
context = toctitle.find("h2").get_text()
```

```python
target = page.find("div", class_ = "toc").find("ul").select("li a")
```

```python
for i in target:
    print(i.get_text())
```
