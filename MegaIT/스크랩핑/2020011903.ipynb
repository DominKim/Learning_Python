```python
# https://en.wikipedia.org/wiki/Kevin_Bacon
from urllib.request import urlopen
from bs4 import BeautifulSoup as bts
import re
import csv

html = urlopen("https://en.wikipedia.org/wiki/Kevin_Bacon")
# https://en.wikipedia.org + /wiki/New_Haven,_Connecticut
bs = bts(html.read(), "lxml")
# 1. 데이터 수집
t_table = bs.find("table", {"class" : "wikitable"})
print(t_table)

rows = t_table.find_all("tr")
len(rows)

# csv 파일 생성
wikitable = open("/Users/mac/Desktop/bigdata/sample/wikitalbe.csv", "wt+", encoding = "UTF-8") # open(제목, wt+ : 읽기 쓰기 모드)
writer = csv.writer(wikitable)

for row in rows:
    cols = row.find_all(["th", "td"])
    wikitable_row = []
    for col in cols:
        wikitable_row.append(col.text.replace("\n", ""))
    print(wikitable_row)
    writer.writerow(wikitable_row)
wikitable.close()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bts
import re
import csv

html = urlopen("http://www.pythonscraping.com/pages/page3.html")
bs = bts(html.read(), "lxml")
# 1. 데이터 수집
giftList = bs.find("table", {"id" : "giftList"})
# print(t_table)

rows = giftList.find_all("tr")

# csv 파일 생성
giftList_csv = open("c:/sample/giftList.csv", "wt+", encoding = "UTF-8") # open(제목, wt+ : 읽기 쓰기 모드)
writer = csv.writer(giftList_csv)

for row in rows:
    cols = row.find_all(["th", "td"])
    giftList_row = []
    for col in cols:
        giftList_row.append(col.text.replace("\n", ""))
    print(giftList_row)
    writer.writerow(giftList_row)
giftList_csv.close() # open -> close를 해야 실행 가능
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bts
import MySQLdb

html = urlopen("http://www.pythonscraping.com/pages/page3.html")
bs = bts(html.read(), "lxml")
# 1. 데이터 수집
giftList = bs.find("table", {"id" : "giftList"})
# print(t_table)
rows = giftList.find_all("tr")
del rows[0]

conn = MySQLdb.connect(host = "localhost", port = 3306, db = "r_db", user = "root", passwd = "python1234")
cursor = conn.cursor()

for row in rows:
    td = row.find_all("td")
    itemtitle = td[0].text.replace("\n","")
    Description = td[1].text.replace("\n","")
    cost = td[2].text.replace("\n","")
    #image = td[3].text
    cursor.execute("insert into mall(itemtitle, Description, cost) values(%s, %s, %s)", [itemtitle, Description, cost])
conn.commit()
conn.close()
```
