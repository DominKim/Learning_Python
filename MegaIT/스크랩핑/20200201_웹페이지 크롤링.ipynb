```python
import requests as rq
from bs4 import BeautifulSoup as bs
import MySQLdb
```

```python
base_url = "https://pjt3591oo.github.io/page"
page = 2

title = []
descript = []
author = []

while True:
    res = rq.get(base_url + str(page))
    if res.status_code != 200:
        break;
    soup = bs(res.content, "lxml", from_encoding= "utf-8")
    # 2, 3, 4 페이지를 크롤링
    posts = soup.select("body main.page-content div.wrapper div.home div.p") # body(부모)태그 안에 (space) main(자식)태그
    for post in posts:
        title.append(post.find("h3").text.strip()) # strip() : 좌우여백삭제
        descript.append(post.find("h4").text.strip())
        author.append(post.find("span").text.strip().replace("\xa0\xa0|\xa0\xa0", "")) # replace("삭제 원하는 문자", "대채문자")
    page += 1
print(len(title))
print(title)
print(len(descript))
print(descript)
print(len(author))
print(author)
```

```python
conn = MySQLdb.connect(host = "localhost", port = 3306, db = "web",
                       user = "root", passwd = "python1234", 
                      use_unicode = True, charset = "utf8") # db 인코딩 강제설정
cursor = conn.cursor()

try:
    sql = "insert into board(title, descript, author) values(%s, %s, %s)"
    for i in range(len(title)):
        print(i)
        print(title[i])
        print(descript[i])
        print(author[i])
        cursor.execute(sql, [title[i], descript[i], author[i]])
except e:
    conn.rollback()
else:
    conn.commit()
finally:
    conn.close()
```

```python
# DB -> 데이터 프레임
import pandas as pd

conn = MySQLdb.connect(host = "localhost", port = 3306, db = "web",
                       user = "root", passwd = "python1234", 
                      use_unicode = True, charset = "utf8") # db 인코딩 강제설정
cursor = conn.cursor()
title = []
descript = []
author = []
try:
    sql = "select title, descript, author from board"
    cursor.execute(sql)
    rows = cursor.fetchall()
    for row in rows:
        # print(row)
        title.append(row[0])
        descript.append(row[1])
        author.append(row[2])
except e:
    print(e)
finally:
    conn.close()

print(title)
print(descript)
print(author)

df = pd.DataFrame({"title" : title, "descript" : descript, "author" : author})
df
```
