```python
# 지도학습 - 1. 회귀분석 : 실수값(연속된 값)을 예측
# from sklearn import datasets
from sklearn.datasets import load_boston
import pandas as pd # 데이터 프레임으로 시각화
import matplotlib.pyplot as plt # 차트 시각화
import seaborn as sns # 차트 시각화
from sklearn.linear_model import LinearRegression # 선형회귀 알고리즘
```

```python
boston = load_boston()
dir(boston)
```

```python
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
df["MEDV"] = boston.target
df.tail()
```

```python
# 차트 시각화 작업 -> 전처리
sns.pairplot([df["RM"], df["MEDV"]])
plt.show()
```

```python
sns.pairplot(df[["MEDV", "RM", "CHAS"]])
plt.show()
```

```python
df.tail()
```

```python
# 모델결정
model = LinearRegression()
# 결정된 모델에 훈련 데이터를 입력해서 훈련
result = model.fit(boston.data, boston.target)
prediction = model.predict(boston.data)
print(prediction)
print(boston.target)
plt.scatter(boston.target, prediction) # 예측을 제대호 한지 확인
plt.show()
```

```python
dir(result)
print(result.intercept_)
print(result.coef_)
```
