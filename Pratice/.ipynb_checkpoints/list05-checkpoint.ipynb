```python
# 리스트 삭제하기(함수활용(
# 방법 1) 값으로 삭제하기
# remove(값) 함수 활용
lst = [10, 20, 30, 40, 50]
lst.remove(20)
print(lst)

lst = [10, 20, 30, 40, 50]
del lst[0]
print(lst)
```

```python

lst = []
check = 0
while True:
    print(lst)
    print("1. 추가하기")
    print("2. 삭제하기")
    print("3. 삽입하기")
    print("4. 종료하기")
    
    menu = int(input("메뉴 선택 : "))
    if menu == 1:
        num = int(input("추가할 숫자 입력 : "))
        lst.append(num)
    elif menu == 2:
        del_num = int(input("삭제할 숫자 입력 : "))
        for i in lst:
            if del_num == i:
                lst.remove(del_num)
            else:
                check = -1
        if check == -1:
            print("잘못입력")
    elif menu == 3:
        insert_idx = int(input("추가할 숫자의 위치 : "))
        if insert_idx <= len(lst):
            insert_num = int(input("추가할 숫자 :"))
            lst.insert(insert_idx, insert_num)
        else:
            print("잘못된 위치 입니다.")
            continue
    else:
        break
```
