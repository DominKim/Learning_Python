```python
'''
1. news Crawling
    url : http://media.daum.net
2. pickle save
    binary file save
'''
import urllib.request as req
from bs4 import BeautifulSoup as bts
url = "http://media.daum.net"
```

```python
# 1. url 
res = req.urlopen(url)
src = res.read()
print(src)
```

```python
# 2. html 
src = src.decode("utf-8")
html = bts(src, "html.parser")
print(html)
```

```python
# 3. tag[ =""] -> "a[class="link_txt"]"
links = html.select("a[class = 'link_txt']")
print(len(links)) # 62
print(links)

crawling_data = [] #  list

for link in links:
    link_str = str(link.string) #  
    crawling_data.append(link_str.strip()) #    
    
print(crawling_data) # list
print(len(crawling_data)) # 62
```

```python
# 4. pickle file save
import pickle

# save
file = open("../data/news_crawling.pickle", mode = "wb")
pickle.dump(crawling_data, file)
print("pickle file saved")
```
